```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("FM_housing.csv",encoding='latin1')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df = df.drop(['Lake'],axis=1)
df = df.drop(['Total Half Baths'], axis=1)
df = df.drop(['Total 3/4 Baths'], axis=1)
               
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Style'] = df['Style'].replace({'Bi Level': '2 Story', '4 Level': '4 Story', '3 Level': '3 Story'})
```

```python
df.nunique()
```

```python
plt.figure(figsize=(20, 20))
df.hist()
plt.subplots_adjust(hspace=1.0, wspace=1.0)
```

```python
df.to_csv("fm_housing_clean.csv", index=False)
```
